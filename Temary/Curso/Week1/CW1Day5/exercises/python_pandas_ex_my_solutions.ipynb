```python
# Import libraries:

import pandas as pd

key_url = 'http://winterolympicsmedals.com/medals.csv'

def prepost(*args, **kwargs):
    df = None
    if 'url' in kwargs:
        df = pd.read_csv(kwargs['url'])

    def inner(func):        ## Cuando llega aquí, se va a buscar la función del decorador
        return func

    if df is not None:
        df.hist()
    return inner

@prepost(url=key_url)
def _f_protected():
    test_value_five = lambda x: True if x > 5 else False
    l1 = [i for i in range(16)]
    return filter(test_value_five, l1)

filter_ = _f_protected()
print(list(filter_))
```
