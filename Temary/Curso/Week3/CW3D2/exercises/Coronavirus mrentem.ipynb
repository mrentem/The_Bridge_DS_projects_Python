## Materiales para analizar series temporales

- Nota: Ejemplo con dataset de consumo de energía en Alemania: https://www.dataquest.io/blog/tutorial-time-series-analysis-with-pandas/

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(8, 4)})
```

```python
df_spain['new_cases'].plot(linewidth=0.5)
plt.title('COVID-19 Spain New cases Reported (Feb 1st 2020 - March 15 2020)')
plt.xlabel('Date')
plt.ylabel('Amount of New Cases')
plt.show()
```

```python
df_spain['total_cases'].plot(linewidth=0.5)
plt.title('COVID-19 Spain Total cases Reported (Feb 1st 2020 - March 15 2020)')
plt.xlabel('Date')
plt.ylabel('Amount of Total Cases')
plt.ylim(0,7000)
plt.show()
```
