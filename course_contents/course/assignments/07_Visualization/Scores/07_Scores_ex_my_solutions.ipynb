# Scores

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

```python
# Preparig raw_data:

raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'age': [42, 52, 36, 24, 73],
            'female': [0, 1, 1, 0, 1],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
```

```python
students = pd.DataFrame(data=raw_data)
students
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
x = students.preTestScore
y = students.postTestScore

plt.scatter(x, y, s=students.age)
plt.title('Students Pre and Post Test Scores')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
x = students.preTestScore
y = students.postTestScore

plt.scatter(x, y, s=(students.postTestScore)*4.5, c=students.female, cmap='cool')
plt.title('Students Pre and Post Test Scores')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.xlim(0,35)
plt.ylim(0,100)
plt.show()
```
