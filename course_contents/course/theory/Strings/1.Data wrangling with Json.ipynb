## Working on Jsons

```python
import requests # con request accedemos a la url
import json # libreria que permite traernos el objeto en formato JSON
import pandas as pd
```

```python
r = requests.get(url='http://citibikenyc.com/stations/json')
json_readed = r.json()
```

```python
print(json_readed)
```

Save json in a local file called "data.json"

```python
# De este modo abrimos el archivo en formato escritura 'w+', y tendra formato outfile.
# El metodo 'dump' crea el archivo json. Hacemos un dump que es volcar ese contenido en el archivo ourfile.

with open('data.json', 'w+') as outfile:
    json.dump(json_readed, outfile)
```

```python
# Con este codigo leemos el archivo linea a linea. Si no esta vacio, lee la linea. Pero esto no es un while, de manera que aunque por logica solo deberia leer una linea, pero como el documento no es un archivo dividido por diferentes lineas. Aqui una solo linea contiene toda la informacion del json. Tenemos que poner lo datos en un formato especifico para poder tratarlos.abs

# OBJETIVO PRINCIPAL: vamos a dividirlo en diferentes lineas.

with open('data.json') as f:
    line = f.readline()
    if line:
        print(line)
        line = f.readline()
print(line)
```

Save with indent

```python
# Por eso ahora vamos a hacer un dump, pero con un indent=4.
# Asi ahora podemos leer el archivo json de una forma mas estructurada --- abrir el archivo.

with open('data_indented.json', 'w+') as outfile:
    json.dump(json_readed, outfile, indent=4)
```

Read local json


```python
with open('data_indented.json', 'r+') as outfile:
    json_readed = json.load(outfile)
json_readed
```

```python
# Podemos leerlo directamente con Pandas read_json, el problema es que lo lee con dos columnas, pero los atributos de cada una de las observaciones esta realmente en stationBeanList.

df_json = pd.read_json("data_indented.json")
df_json
```
