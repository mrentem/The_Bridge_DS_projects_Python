```python
## Generamos una palabra aleatoriamente utilizando el diccionation SOWPODS:

  #Importamos random

import random

  # read all the list of words
words = []
with open('sowpods.txt', 'r') as f:
    line = f.readline().strip()
    words.append(line)
    while line:
        line = f.readline().strip()
        words.append(line)

  # generate a random number
    random_index = random.randint(0, len(words))

  # take the word
    adivinar = words[random_index]
    #print("Random word: ", adivinar)


SPACE = " "     # Las variable globales, generalmente se definen en mayuscula
LOW_BAR = "_"


## Creamos el tablero con el número de letras de la palabra a adivinar y un segundo tablero que es el que le vamos a mostrar al jugador para que sepa cuántas letras tiene que adivinar:

def create_board(word): # Esta función crea el tablero que vamos a comparar con las letras que diga el jugador
    board =""
    for i in range(len(word)):
        board += LOW_BAR
    return board

create_board(word=adivinar)

## Comprobar si una letra está en una palabra:

def check_if_exists(letter, word):
    
    if letter in word:
        return True
    else:
        return False

    #return letter in word --> haría el mismo efecto

def get_indexes(letter, word):
    indexes = [] # Lista de los índices en los que coincide la letra

    for i, e in enumerate(word):
        if e == letter:
            indexes.append(i)  # Metemos en la lista de índices aquellos en los que la letra dada es igual a la de la palabra
    return indexes

def show_letters_in_board(board, indexes, letter):
    
    for index, elemento in enumerate(board):
        if index in indexes:
            board[index] = letter
    show_string_board(board)
    return board


def show_string_board(board):
    board_to_show = ""
    for x in board:
        board_to_show += x + SPACE   # En el tablero que le mostremos al jugador incluiremos un espacio para que sepa las letras que tiene
    print(board_to_show)


def start_guess_letter():

    # Step 0: crear la palabra -- ir a la parte de arriva
    adivinar = words[random_index]

    # Step 1: crear el tablero
    board = list(create_board(word=adivinar)) # Tenemos que crearlo como lista para poder recorrerlo y sustituir las letras
    show_string_board(board=board)
    
    while True:

        # Step 2: Introducir la letra
        user_letter = input("Introduce una letra:").upper() #Necesitamos que sea en mayúsculas
        if user_letter == "STOP": # Damos un escape en caso de que el jugador no quiera seguir
            print("Eres un flojo. La palabra era:", adivinar)
            break
        
        # Step 3: comprobar si la letra existe:
        existe_letra = check_if_exists(letter=user_letter, word=adivinar)
        
        # Step 4: Si existe, mostrar letra en el tablero
        if existe_letra:
            indexes = get_indexes(letter=user_letter, word=adivinar)
            board = show_letters_in_board(board=board, indexes=indexes, letter=user_letter) # Nuestro board ha cambiado
            if not "_" in board:    # Ha acertado todas y ya no quedan "_"
                print("Eres un campeón")
                break
        else:
            print("No está esa letra")


start_guess_letter()
```
